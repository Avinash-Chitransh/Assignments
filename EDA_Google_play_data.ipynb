```python
def new_size (item):
  if str(item)[-1] == "M":
    re_size = float(str(item).replace("M", ""))
    re_size = re_size * 1000
    return re_size
  elif str(item)[-1] == "k":
    re_size = float(str(item).replace("k", ""))
    return re_size
  else:
    return (np.nan)
```

```python
def extract_major_minor(val):
    try:
        if pd.isnull(val):
            return np.nan
        parts = str(val).split('.')
        if len(parts) >= 2:
            return float(f"{parts[0]}.{parts[1]}")
        elif len(parts) == 1:
            return float(parts[0])
        else:
            return np.nan
    except:
        return np.nan

df['Current Ver'] = df['Current Ver'].apply(extract_major_minor)
```

```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

# Step 1: Select only numeric columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Step 2: Drop rows with NaN values
numeric_df = numeric_df.dropna()

# Step 3: Add a constant for intercept (required by statsmodels)
X = add_constant(numeric_df)

# Step 4: Calculate VIF for each variable
vif_data = pd.DataFrame()
vif_data["Feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

# Step 5: Drop constant row (optional, since it’s not a feature)
vif_data = vif_data[vif_data["Feature"] != "const"]

# Display
print(vif_data)
```
