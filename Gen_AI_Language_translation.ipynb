```python
from transformers import pipeline

# Create a translation pipeline for English to Hindi
translator = pipeline("translation_en_to_hi", model="Helsinki-NLP/opus-mt-en-hi")

# Example text in English
text = "How are you?"

# Translate English text to Hindi
translated = translator(text)

# Print the translated text
print(translated)
```

```python
translator = pipeline("translation_en_to_jap", model="Helsinki-NLP/opus-mt-en-jap")

# Example text in English
text_1 = "How are you?"

# Translate English text to Japanese
translated_1 = translator(text_1)

# Print the translated text
print(translated_1)
```

```python
translator = pipeline("translation_en_to_es", model="Helsinki-NLP/opus-mt-en-es")

# Example text in English
text_2 = "How are you?"

# Translate English text to Spanish
translated_2 = translator(text_2)

# Print the translated text
print(translated_2)
```
