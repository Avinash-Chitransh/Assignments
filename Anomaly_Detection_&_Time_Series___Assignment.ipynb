1. What is Anomaly Detection? Explain its types (point, contextual, and
collective anomalies) with examples.

Anomaly detection is the process of identifying data patterns or instances that deviate significantly from the expected behavior of a dataset. These unusual observations are called anomalies or outliers. Detecting anomalies is crucial in various domains such as fraud detection in banking, fault detection in mechanical systems, medical diagnosis, and cybersecurity. Since anomalies often signify critical or unexpected events, their identification helps in early warnings, risk reduction, and decision-making.

Unlike regular data mining or classification, anomaly detection focuses not on general trends but on the rare, unusual events. The challenge arises because anomalies are usually sparse, context-dependent, and may not follow consistent patterns. To systematically study them, anomalies are classified into three main types: point anomalies, contextual anomalies, and collective anomalies.

1. Point Anomalies

A point anomaly occurs when an individual data point is significantly different from the rest of the data. These are the simplest and most common form of anomalies. Point anomalies often indicate direct errors in measurement, data entry mistakes, or rare but valid occurrences.

Example 1 (Banking): If a customer usually withdraws between ₹5,000 and ₹10,000 monthly but suddenly withdraws ₹1,00,000 in a single transaction, that transaction is a point anomaly.

Example 2 (Healthcare): In patient monitoring, if a person’s normal body temperature is around 37°C but one reading shows 42°C, it signals a possible error or medical emergency.

Point anomalies are relatively straightforward to detect using statistical thresholds (e.g., z-scores, boxplots), distance-based methods, or clustering.

2. Contextual Anomalies

A contextual anomaly arises when a data point is anomalous in a specific context but may appear normal in another. Contextual anomalies depend heavily on time, location, or surrounding circumstances. Unlike point anomalies, they cannot be identified without considering the context.

Example 1 (Weather): A temperature of 30°C is normal in summer but anomalous in winter in the same location.

Example 2 (Network Traffic): High server requests at 9 a.m. during work hours may be normal, but the same traffic volume at 3 a.m. could signal an intrusion attempt.

Detecting contextual anomalies usually requires time-series analysis, seasonal decomposition, or models that incorporate domain-specific context, such as seasonality and trends.

3. Collective Anomalies

A collective anomaly occurs when a group of data points together is anomalous, even if individual points may appear normal. These anomalies typically arise in sequence or batch data, where patterns across multiple points indicate abnormality.

Example 1 (Cybersecurity): A sudden surge of small but repeated login attempts from the same IP address over time could indicate a brute-force attack. Each attempt may seem harmless, but collectively they form a suspicious pattern.

Example 2 (Healthcare): A patient’s heart rate readings might individually fall within acceptable ranges, but a series of irregular fluctuations over an hour may reveal an abnormal heart condition.

Collective anomalies are often detected using sequence analysis, clustering, and machine learning models such as Hidden Markov Models (HMMs) or Recurrent Neural Networks (RNNs).

Anomaly detection plays a vital role in identifying rare but meaningful events hidden within large datasets. Point anomalies highlight individual outliers, contextual anomalies expose irregularities relative to their environment, and collective anomalies reveal unusual patterns across multiple data points. By distinguishing these categories, organizations can build targeted detection systems—whether for fraud monitoring, system reliability, or medical safety. As data grows increasingly complex, advanced statistical models, machine learning, and deep learning methods are being employed to enhance anomaly detection, ensuring timely intervention and greater accuracy.

2. Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.

a. Isolation Forest (iForest)

Approach:
Isolation Forest is an ensemble-based anomaly detection algorithm that works by isolating data points instead of profiling normal behavior. It constructs random decision trees where anomalies require fewer splits to isolate because they lie far from dense regions. The anomaly score is based on the average path length across many trees.



Strengths:

*   Efficient for high-dimensional datasets.
*   Works well with large-scale data due to linear time complexity.
*   Does not assume any underlying data distribution.

Limitations:

*   Less effective for datasets with local density variations.
*   Sensitive to contamination (expected anomaly proportion) parameter.
*   Suitable Use Cases:
    Fraud detection in banking and e-commerce.

    Network intrusion detection.

    High-dimensional data such as text or genetic datasets.


b. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

Approach:
DBSCAN is a density-based clustering algorithm that groups together points in dense regions and labels points in sparse regions as outliers. It relies on two parameters: ε (neighborhood radius) and minPts (minimum number of points required to form a dense region). Anomalies are detected as points not belonging to any cluster.

Strengths:
*   Can discover arbitrarily shaped clusters.
*   Handles noise and outliers naturally as part of its clustering process.
*   No need to specify the number of clusters beforehand.

Limitations:

*   Struggles with varying density clusters (choice of ε is critical).

*   Not scalable to very high-dimensional data.

*   Suitable Use Cases:

    Geospatial anomaly detection (e.g., unusual GPS locations).

    Image segmentation or clustering of irregular patterns.

    Detecting anomalies in low-to-medium dimensional datasets with spatial structure.


c. Local Outlier Factor (LOF)

Approach:
LOF is a density-based local anomaly detection method. It measures the local density deviation of a data point relative to its neighbors. A point is considered an anomaly if its density is significantly lower than that of its neighbors. LOF computes an anomaly score (LOF score), where values much greater than 1 indicate anomalies.

Strengths:


*   Effective in datasets with local density variations.

*   Considers relative density, making it good for non-uniform data distributions.
*   Does not require assumptions about global data distribution.

Limitations:


*   Computationally expensive for large datasets (distance calculations).
*   Sensitive to parameter k (number of neighbors).
*   Struggles in high-dimensional data due to curse of dimensionality.
*   Suitable Use Cases:

    Credit card fraud detection (small local deviations matter).

    Sensor data analysis (detecting localized irregular patterns).

    Detecting anomalies in datasets with clusters of different densities.













3. What are the key components of a Time Series? Explain each with one example.

Key Components of a Time Series

A time series is a sequence of data points collected at successive time intervals, often equally spaced (e.g., daily stock prices, monthly rainfall, yearly GDP). To properly analyze and forecast time series data, it is broken down into its four key components: Trend, Seasonality, Cyclical Variation, and Irregular/Random Variation.

1. Trend (T)

Definition:
The long-term movement or direction in the data over a period of time. It represents whether the data is generally increasing, decreasing, or remaining stable.

Example:
The number of internet users worldwide shows an upward trend over the last two decades as more people gain access to technology.

2. Seasonality (S)

Definition:
Regular, predictable patterns in data that repeat over fixed intervals (such as daily, weekly, monthly, or yearly). Seasonality is caused by factors such as weather, holidays, or cultural practices.

Example:
Ice cream sales rise every summer and drop during winter. This repeating yearly pattern is seasonality.

3. Cyclical Variation (C)

Definition:
Long-term fluctuations in the data that occur over periods longer than one year. Unlike seasonality, cycles are not fixed or regular. They are often influenced by macroeconomic or business factors.

Example:
The business cycle, with periods of economic expansion followed by recession, reflects cyclical variation in GDP data. Unlike seasonal changes, these cycles do not follow a strict calendar pattern.

4. Irregular or Random Variation (I)

Definition:
Unpredictable, short-term fluctuations in the data caused by unforeseen events such as natural disasters, strikes, accidents, or sudden policy changes. These variations cannot be modeled easily.

Example:
A sudden drop in airline travel during the COVID-19 pandemic is an irregular variation, as it was caused by an unexpected global crisis.


Trend shows the long-term direction,

Seasonality reflects short-term, regular patterns,

Cyclical variation captures long-term but irregular fluctuations, and

Irregular variation represents unexpected disturbances.

By decomposing a time series into these components, analysts can better understand data behavior and improve forecasting accuracy.


4. Define Stationary in time series. How can you test and transform a
non-stationary series into a stationary one?

A time series is said to be stationary if its statistical properties—such as mean, variance, and autocovariance—remain constant over time. In other words, the behavior of the series does not depend on the time at which it is observed. Stationarity is important because most time series models (like ARIMA) assume the input data is stationary; otherwise, the forecasts may be misleading.

Example of Stationary Series: Daily fluctuations in temperature around a fixed mean for one month.

Example of Non-Stationary Series: Stock prices, which generally exhibit trends and changing variance.

Types of Stationarity

Strict Stationarity: The complete distribution of the series is time-invariant.

Weak (or Covariance) Stationarity: Only mean, variance, and autocovariance remain constant over time. This weaker form is often assumed in practice.

Testing for Stationarity

Several statistical and visual methods are used:

Visual Inspection:

Plot the series. If it shows a clear trend, seasonality, or changing variance, it is likely non-stationary.

Summary Statistics over Time:

Divide the series into different time windows and check if mean and variance are stable.

Statistical Tests:

Augmented Dickey-Fuller (ADF) Test: Null hypothesis assumes the series is non-stationary (unit root present).

KPSS Test: Null hypothesis assumes the series is stationary.

Phillips-Perron (PP) Test: Another alternative to test for unit roots.

Transforming a Non-Stationary Series into a Stationary One

Differencing:

Subtract the previous observation from the current observation.

Example: If series shows a trend, applying first-order differencing (Yₜ – Yₜ₋₁) often removes it.

Seasonal differencing (Yₜ – Yₜ₋ₘ) removes seasonality, where m is the seasonal period.

Transformation (Variance Stabilization):

Apply logarithm, square root, or Box-Cox transformation to stabilize variance.

Example: Log-transforming stock prices to reduce volatility.

Detrending:

Fit a regression line or polynomial to remove the trend component, leaving behind stationary residuals.

Seasonal Adjustment:

Use decomposition methods to remove the seasonal component before modeling.


Stationarity ensures that statistical inferences and forecasting remain valid over time. A non-stationary series can often be made stationary through differencing, transformation, detrending, or seasonal adjustment, after confirming non-stationarity via tests such as ADF or KPSS. Once transformed, models like ARIMA can be effectively applied.

```python
# Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies. Visualize the anomalies on a 2D scatter plot.

from sklearn.ensemble import IsolationForest
```

```python
# Step 1: Create a synthetic Taxi dataset
np.random.seed(42)
# Normal trips: trip_distance (km) and fare_amount ($)
trip_distance = np.random.normal(5, 2, 300)   # avg 5 km, std 2
fare_amount = trip_distance * 3 + np.random.normal(0, 2, 300)  # ~ $3 per km
```

```python
#add anamolies
anamoly_distance = np.random.normal(10, 2, 20)
anamoly_fare = anamoly_distance * 3 + np.random.normal(0, 2, 20)
```

```python
final_distance = np.concatenate((trip_distance, anamoly_distance))
final_fare = np.concatenate((fare_amount, anamoly_fare))
```

```python
df = pd.DataFrame({'trip_distance': final_distance, 'fare_amount': final_fare})
```

```python
df.head()
```

```python
model = IsolationForest(contamination=0.1, random_state=42, n_jobs=-1)
```

```python
df['anamoly'] = model.fit_predict(df[['trip_distance', 'fare_amount']])
```

```python
df['anamoly']
```

```python
import matplotlib.patches as mpatches
plt.figure(figsize=(5,5))
plt.scatter(df['trip_distance'], df['fare_amount'], c=df['anamoly'], cmap='coolwarm', edgecolors="k")
plt.xlabel('trip_distance(km)')
plt.ylabel('fare_amount($)')
plt.title('Isolation Forest Anomalies')
plt.tight_layout()
normal_patch = mpatches.Patch(color='blue', label='Normal')
anomaly_patch = mpatches.Patch(color='red', label='Anomaly')
plt.legend(handles=[normal_patch, anomaly_patch])
plt.show()
```

```python
# Train a SARIMA model on the monthly airline passengers dataset. Forecast the next 12 months and visualize the results.

from statsmodels.datasets import get_rdataset
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
data = get_rdataset("AirPassengers").data
```

```python
data.head()
```

```python
df = data.rename(columns={"value": "passengers"})
df['Month'] = pd.date_range(start='1949-01-01', periods=len(df), freq='MS')
df.set_index('Month', inplace=True)
```

```python
df.head()
```

```python
model = SARIMAX(df['passengers'],
                order=(1,1,1),
                seasonal_order=(1,1,1,12),
                enforce_stationarity=False,
                enforce_invertibility=False)

results = model.fit()
```

```python
# Forecast next 12 months
forecast = results.get_forecast(steps=12)
forecast_ci = forecast.conf_int()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df.index, df['passengers'], label='Observed')
plt.plot(forecast.predicted_mean.index, forecast.predicted_mean,
         label='Forecast', color='red')
```

```python
# Confidence interval shading
plt.fill_between(forecast_ci.index,
                 forecast_ci.iloc[:, 0],
                 forecast_ci.iloc[:, 1], color='pink', alpha=0.3)

plt.title("SARIMA Forecast of Airline Passengers")
plt.xlabel("Date")
plt.ylabel("Number of Passengers")
plt.legend()
plt.show()
```

```python
#Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib.

from sklearn.neighbors import LocalOutlierFactor
from sklearn.datasets import load_diabetes
```

```python
data = load_diabetes()
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
df.head()
```

```python
x = df[['bmi', 'bp']]
```

```python
x.head()
```

```python
# Apply Local Outlier Factor
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)  # 5% anomalies
y_pred = lof.fit_predict(x)
```

```python
y_pred
```

```python
# -1 = anomaly, 1 = normal
df['anomaly'] = y_pred
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x['bmi'], x['bp'],
            c=df['anomaly'], cmap='coolwarm', edgecolor="k", s=60)

plt.title("Local Outlier Factor (LOF) Anomaly Detection")
plt.xlabel("bmi")
plt.ylabel("bp)")
plt.legend(handles=[
    plt.Line2D([0], [0], marker='o', color='w', label='Normal',
               markerfacecolor='blue', markersize=10, markeredgecolor="k"),
    plt.Line2D([0], [0], marker='o', color='w', label='Anomaly',
               markerfacecolor='red', markersize=10, markeredgecolor="k")
])
plt.show()
```

1. Detecting Anomalies in Streaming Data

Since the dataset arrives every 15 minutes, anomaly detection must be online and scalable. Options:

Isolation Forest:

Works well for high-dimensional data (e.g., energy + weather + region).

Detects unusual spikes/drops that deviate from normal distribution.

Suitable for batch windows (e.g., sliding 1-hour or 24-hour window).

Local Outlier Factor (LOF):

Captures local density variations.

Useful if anomalies are relative to neighboring points (e.g., one region consumes abnormally high compared to others under similar weather).

DBSCAN:

Clusters normal patterns, flags points outside clusters as anomalies.

Works if consumption patterns form natural clusters (e.g., peak vs. off-peak hours).

Less suitable for high-velocity streaming due to higher computation cost.

 Recommended approach: Isolation Forest (for global anomalies) + LOF (for local anomalies across regions). Both can be implemented in mini-batch streaming mode.

🔹 2. Short-Term Forecasting Model

For 15-minute interval energy usage:

ARIMA: Captures trend + seasonality but assumes stationarity. Works for univariate demand only.

SARIMA: Adds explicit seasonality terms → good for daily/weekly demand cycles.

SARIMAX: Adds exogenous variables (e.g., weather, temperature, holidays, region). Best for real-world energy forecasting.

LSTM/GRU (Deep Learning): Handles nonlinearities, long-term dependencies, large-scale data better.

 Recommended: SARIMAX (for interpretable forecasts using weather & region features) or LSTM (if business prioritizes accuracy over interpretability).

🔹 3. Validation & Monitoring

Validation methods:

Rolling Forecast Origin (walk-forward validation) → ensures real-time simulation.

Metrics: RMSE, MAPE (for forecast error), Precision/Recall (for anomaly detection).

Monitoring over time:

Drift detection → monitor if model residuals increase (concept drift due to new usage patterns).

Retraining schedule → update SARIMAX weekly or use online learning models.

Real-time dashboards → Streamlit / Grafana to visualize demand vs. forecast and anomalies flagged.

🔹 4. Business Impact

This solution helps in:

Grid Stability: Detect abnormal spikes (possible equipment failure, fraud, or sudden demand surge) before outages occur.

Cost Optimization: Forecasting allows better planning of energy purchase and generation schedules.

Demand Response: Utilities can trigger load-balancing strategies or incentives for consumers during peak loads.

Operational Efficiency: Reduces downtime, improves regulatory compliance, and enhances customer satisfaction.

Final Workflow:

Streaming data → Preprocessing & feature engineering (region, weather, time-of-day).

Anomaly detection → Isolation Forest + LOF on mini-batches.

Forecasting → SARIMAX for 12–48 hour short-term demand prediction.

Monitoring → Rolling validation, retraining schedule, dashboard alerts.

Business → Prevent blackouts, optimize costs, improve decision-making.
