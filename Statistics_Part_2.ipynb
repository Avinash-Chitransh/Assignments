```python
import scipy.stats as stats
import numpy as np

def margin_of_error(sample_data, confidence=0.95):
    n = len(sample_data)
    mean = np.mean(sample_data)
    std_err = stats.sem(sample_data)
    margin = std_err * stats.t.ppf((1 + confidence) / 2, n - 1)
    return margin

data = [23, 25, 27, 24, 26, 22, 28, 25, 24, 26]
moe = margin_of_error(data, confidence=0.95)
mean = np.mean(data)
print(f"Mean: {mean:.2f} ± {moe:.2f}")
```

```python
def bayes_theorem(p_b_given_a, p_a, p_b):
    return (p_b_given_a * p_a) / p_b
print("Bayesian Inference:", bayes_theorem(0.8, 0.5, 0.6))
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

data = {'Gender': ['M', 'F', 'M', 'F', 'M', 'F'], 'Preference': ['A', 'B', 'A', 'A', 'B', 'B']}
df = pd.DataFrame(data)
def chi_square_test(df):
    table = pd.crosstab(df.iloc[:,0], df.iloc[:,1])
    chi2, p, dof, ex = chi2_contingency(table)
    return chi2, p
print("Chi-square Test:", chi_square_test(df))
```

```python
import numpy as np
from scipy import stats

# Sample data: test scores of students
sample_scores = np.array([78, 74, 69, 81, 77, 72, 79, 85, 70, 76])

# Null hypothesis: population mean = 75
population_mean = 75

# Perform one-sample t-test
t_statistic, p_value = stats.ttest_1samp(sample_scores, population_mean)

# Output the results
print("T-statistic:", round(t_statistic, 3))
print("P-value:", round(p_value, 4))

# Interpret the result
alpha = 0.05  # significance level
if p_value < alpha:
    print("Reject the null hypothesis: significant difference from the population mean.")
else:
    print("Fail to reject the null hypothesis: no significant difference from the population mean.")
```

```python
sample1 = [10, 20, 30, 40, 50]
sample2 = [15, 25, 35, 45, 55]
def f_test(var1, var2):
    f = np.var(var1, ddof=1) / np.var(var2, ddof=1)
    dfn = len(var1) - 1
    dfd = len(var2) - 1
    p = 1 - stats.f.cdf(f, dfn, dfd)
    return f, p
print("F-test:", f_test(sample1, sample2))
```

```python
g1 = [23, 45, 67]
g2 = [12, 34, 56]
g3 = [14, 27, 31]
from scipy.stats import f_oneway
def anova_test(*groups):
    return f_oneway(*groups)
print("ANOVA Test:", anova_test(g1, g2, g3))
```

```python
def anova_with_plot(groups):
    f_val, p_val = f_oneway(*groups)
    plt.boxplot(groups)
    plt.title("One-way ANOVA")
    plt.show()
    return f_val, p_val
print("One-way ANOVA:", anova_with_plot([g1, g2, g3]))
```

```python
from scipy.stats import shapiro, levene

def check_anova_assumptions(groups):
    normality = [shapiro(group)[1] > 0.05 for group in groups]
    _, p_val = levene(*groups)
    equal_var = p_val > 0.05
    return all(normality), equal_var
print("Assumptions Check:", check_anova_assumptions([g1, g2, g3]))
```

```python
import numpy as np
from scipy.stats import chi2

def variance_test(data, sigma2, alpha=0.05):
    n = len(data)
    sample_var = np.var(data, ddof=1)  # unbiased sample variance
    chi2_stat = (n - 1) * sample_var / sigma2

    # Two-tailed test: compute critical region
    chi2_lower = chi2.ppf(alpha / 2, df=n - 1)
    chi2_upper = chi2.ppf(1 - alpha / 2, df=n - 1)

    # P-value calculation (two-tailed)
    p_value = 2 * min(
        chi2.cdf(chi2_stat, df=n - 1),
        1 - chi2.cdf(chi2_stat, df=n - 1)
    )

    result = "Reject H₀" if chi2_stat < chi2_lower or chi2_stat > chi2_upper else "Fail to Reject H₀"

    return {
        "Chi2 Statistic": round(chi2_stat, 4),
        "P-value": round(p_value, 4),
        "Decision": result
    }

# Example usage
data = np.random.normal(loc=60, scale=8, size=20)
print("Variance Test:", variance_test(data, sigma2=64))
```

```python
from math import sqrt
from scipy.stats import norm

def z_test_proportions(p1, p2, n1, n2, alpha=0.05):
    # Number of successes
    x1 = p1 * n1
    x2 = p2 * n2

    # Pooled proportion
    p_hat = (x1 + x2) / (n1 + n2)

    # Standard error
    se = sqrt(p_hat * (1 - p_hat) * (1/n1 + 1/n2))

    # Z-score
    z = (p1 - p2) / se

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z)))

    # Decision
    result = "Reject H₀" if p_value < alpha else "Fail to Reject H₀"

    return {
        "Z-score": round(z, 4),
        "P-value": round(p_value, 4),
        "Decision": result
    }

# Example usage
print("Z-test Proportions:", z_test_proportions(0.6, 0.5, 100, 100))
```

```python
sample = np.random.normal(52, 5, 30)
pop_mean = 50
def z_test_sample_mean(sample, pop_mean):
    z = (np.mean(sample) - pop_mean) / (np.std(sample, ddof=1) / np.sqrt(len(sample)))
    p = 2 * (1 - stats.norm.cdf(abs(z)))
    return z, p
print("Z-test Sample Mean:", z_test_sample_mean(sample, pop_mean))
```

```python
sample = np.random.normal(100, 10, 50)
pop_mean = 95
z_stat = (np.mean(sample) - pop_mean) / (np.std(sample, ddof=1) / np.sqrt(len(sample)))
def plot_two_tailed_z(z_stat):
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x)
    plt.plot(x, y)
    plt.fill_between(x, y, where=(x <= -abs(z_stat)) | (x >= abs(z_stat)), color='red', alpha=0.5)
    plt.axvline(z_stat, color='blue')
    plt.title("Two-tailed Z-test")
    plt.show()
plot_two_tailed_z(z_stat)
print("Two-tailed Z-test stat:", z_stat)
```

```python
def visualize_errors():
    x = np.linspace(-4, 4, 1000)
    y0 = stats.norm.pdf(x, 0, 1)
    y1 = stats.norm.pdf(x, 2, 1)
    plt.plot(x, y0, label='Null Hypothesis')
    plt.plot(x, y1, label='Alternative Hypothesis')
    plt.fill_between(x, y0, where=(x > 1.645), color='red', alpha=0.5, label='Type I Error')
    plt.fill_between(x, y1, where=(x <= 1.645), color='blue', alpha=0.3, label='Type II Error')
    plt.legend()
    plt.title("Type I and Type II Errors")
    plt.show()
visualize_errors()
```

```python
sample1 = np.random.normal(50, 5, 30)
sample2 = np.random.normal(55, 5, 30)
def independent_t_test(s1, s2):
    t_stat, p = stats.ttest_ind(s1, s2)
    return t_stat, p
print("Independent T-test:", independent_t_test(sample1, sample2))
```

```python
before = np.random.normal(70, 5, 20)
after = before + np.random.normal(-2, 2, 20)
def paired_t_test(bef, aft):
    t_stat, p = stats.ttest_rel(bef, aft)
    return t_stat, p
print("Paired T-test:", paired_t_test(before, after))
```

```python
def confidence_interval(sample, confidence=0.95):
    mean = np.mean(sample)
    std_err = stats.sem(sample)
    margin = std_err * stats.t.ppf((1 + confidence) / 2, len(sample) - 1)
    return mean - margin, mean + margin
print("Confidence Interval:", confidence_interval(sample))
```
