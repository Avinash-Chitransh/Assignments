1. What is Deep Learning? Briefly describe how it evolved and how it differs from traditional machine learning.

Deep Learning is a branch of machine learning that uses artificial neural networks with many layers (hence "deep") to automatically learn representations of data. It is particularly powerful for complex tasks like image recognition, natural language processing, and speech recognition, where traditional machine learning struggles with feature extraction.

Evolution of Deep Learning

a. 1950s‚Äì1980s (Early AI & Neural Nets): Inspired by the human brain, researchers developed the perceptron (1958). However, limited computing power and the inability of shallow networks to model complex functions led to slow progress.

b. 1990s (Neural revival): Backpropagation was rediscovered, enabling multi-layer neural networks to train more effectively. Yet, training deep models was still computationally expensive.

c. 2000s (Data + GPUs): With the rise of big data and GPU acceleration, researchers could train much larger networks. New architectures like Convolutional Neural Networks (CNNs) and Recurrent Neural Networks (RNNs) solved vision and sequence tasks.

d. 2010s onwards: Breakthroughs like AlexNet (2012) in image recognition and transformers (2017) in NLP established deep learning as the dominant paradigm in AI.

Difference from Traditional Machine Learning

Feature Engineering

Traditional ML: Requires manual feature extraction (e.g., edges in images, statistical descriptors in text).

Deep Learning: Learns hierarchical features automatically from raw data.

Data Requirement

Traditional ML: Performs well with small to medium datasets.

Deep Learning: Requires large datasets to achieve high accuracy.

Model Complexity

Traditional ML: Uses simpler models (decision trees, SVMs, regression).

Deep Learning: Uses multi-layer neural networks with millions of parameters.

Computational Power

Traditional ML: Can run efficiently on CPUs.

Deep Learning: Needs GPUs/TPUs due to high computational demand.

In essence, deep learning evolved from neural networks by leveraging data abundance, advanced algorithms, and modern hardware, making it the driving force behind today‚Äôs AI breakthroughs.

3. Describe the purpose of activation function in neural networks. Compare
Sigmoid, ReLU, and Tanh functions.

The activation function in a neural network plays a crucial role by introducing non-linearity into the model. Without it, a neural network, regardless of how many layers it has, would essentially behave like a linear regression model, unable to capture complex, non-linear relationships in data. Activation functions decide how much of the weighted input signal is passed forward, allowing the network to learn intricate patterns, approximate complex functions, and handle tasks such as image recognition, speech processing, and natural language understanding.

The sigmoid function was one of the earliest activation functions used. It maps input values into a smooth S-shaped curve ranging between 0 and 1, which makes it particularly useful when outputs need to be interpreted as probabilities. However, it suffers from two key limitations: the vanishing gradient problem, where gradients become too small during backpropagation, and the fact that it is not zero-centered, which can slow down learning.

The tanh function (hyperbolic tangent) is similar in shape to sigmoid but outputs values between ‚Äì1 and 1, making it zero-centered. This property generally improves optimization since positive and negative values can cancel out more naturally. Yet, like sigmoid, it still suffers from the vanishing gradient problem in deep networks, limiting its effectiveness for very deep architectures.

The ReLU function (Rectified Linear Unit) has become the most widely used activation function in modern neural networks. It outputs zero for all negative inputs and returns the input value directly for positive inputs. ReLU is computationally simple, efficient, and alleviates the vanishing gradient problem for positive inputs, enabling faster convergence in deep networks. However, it can lead to the issue of ‚Äúdying ReLUs,‚Äù where some neurons stop updating if they consistently receive negative inputs. Variants like Leaky ReLU and ELU were introduced to overcome this limitation.

In essence, sigmoid is best suited for binary outputs, tanh is a better choice than sigmoid for hidden layers when zero-centered activation is helpful, while ReLU and its variants dominate in deep networks due to their efficiency and ability to support gradient-based learning effectively.

5. What is the role of optimizers in neural networks? Compare Gradient Descent, Adam, and RMSprop.

Optimizers in neural networks are algorithms that adjust the weights and biases during training to minimize the cost (or loss) function. Their role is to guide the network‚Äôs learning by updating parameters in the right direction and with the right magnitude, ensuring the model converges efficiently to a solution that generalizes well. Without optimizers, backpropagation would compute gradients, but the network would not know how to use them effectively to improve performance.

The most fundamental optimizer is Gradient Descent. It works by computing the gradient of the cost function with respect to the parameters and updating them in the opposite direction of the gradient. While conceptually simple, plain gradient descent can be slow and may get stuck in local minima or saddle points. Variants like Stochastic Gradient Descent (SGD) improve efficiency by updating parameters using mini-batches rather than the whole dataset, adding randomness that helps escape local minima.

RMSprop (Root Mean Square Propagation) improves on SGD by adapting the learning rate for each parameter individually. It keeps a moving average of the squared gradients and divides the learning rate by the root of this average. This prevents oscillations and helps the optimizer perform better in problems with non-stationary objectives, such as recurrent neural networks.

Adam (Adaptive Moment Estimation) combines the strengths of both RMSprop and momentum. It not only adapts the learning rate for each parameter but also keeps track of exponentially decaying averages of past gradients (first moment) and squared gradients (second moment). This dual adaptation makes Adam faster, more stable, and well-suited for a wide variety of deep learning tasks.

In essence, Gradient Descent is simple but slower, RMSprop adapts learning rates for better handling of complex landscapes, and Adam is generally preferred because it balances speed, stability, and adaptability in training deep neural networks.

6. Write a Python program to implement a single-layer perceptron from
scratch using NumPy to solve the logical AND gate

```python
import numpy as np

# Define the input data and corresponding labels for the AND gate
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
```

```python
y = np.array([0,0,0,1]) #outputs
```

```python
# Initialize parameters
np.random.seed(42)  # for reproducibility
weights = np.random.randn(2)  # two inputs
bias = np.random.randn()
learning_rate = 0.1
epochs = 20
```

```python
# Define activation function (step function)
def step_function(z):
    return 1 if z >= 0 else 0
```

```python
# Training the perceptron
for epoch in range(epochs):
    total_error = 0
    for i in range(len(X)):
        z = np.dot(X[i], weights) + bias
        y_pred = step_function(z)
        error = y[i] - y_pred

        # Update rule
        weights += learning_rate * error * X[i]
        bias += learning_rate * error

        total_error += abs(error)

    print(f"Epoch {epoch+1}, Weights: {weights}, Bias: {bias}, Total Error: {total_error}")

# Step 5: Testing the trained perceptron
print("\nFinal Results:")
for i in range(len(X)):
    z = np.dot(X[i], weights) + bias
    y_pred = step_function(z)
    print(f"Input: {X[i]}, Predicted: {y_pred}, Actual: {y[i]}")
```

```python
print("\nFinal Results:")
correct_predictions = 0
for i in range(len(X)):
    z = np.dot(X[i], weights) + bias
    y_pred = step_function(z)
    print(f"Input: {X[i]}, Predicted: {y_pred}, Actual: {y[i]}")
    if y_pred == y[i]:
        correct_predictions += 1
```

```python
accuracy = correct_predictions / len(X) * 100
print(f"\nAccuracy: {accuracy:.2f}%")
```

7. Implement and visualize Sigmoid, ReLU, and Tanh activation functions
using Matplotlib.


```python
import matplotlib.pyplot as plt

# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)
```

```python
# Generate input values
x = np.linspace(-10, 10, 400)

# Compute function outputs
y_sigmoid = sigmoid(x)
y_relu = relu(x)
y_tanh = tanh(x)
```

```python
plt.figure(figsize=(8, 4))
#sigmoid
plt.subplot(2, 2, 1)
plt.plot(x, y_sigmoid, label='Sigmoid')
plt.title('Sigmoid Activation Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

```python
plt.figure(figsize=(8, 4))
#ReLU
plt.subplot(2, 2, 2)
plt.plot(x, y_relu, label='ReLU')
plt.title('ReLU Activation Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

```python
plt.figure(figsize=(8, 4))
#tanh
plt.subplot(2, 2, 3)
plt.plot(x, y_tanh, label='Tanh')
plt.title('Tanh Activation Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

8. Use Keras to build and train a simple multilayer neural network on the
MNIST digits dataset. Print the training accuracy.


```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
# Preprocess the data
x_train = x_train.reshape(-1, 28*28).astype("float32") / 255.0
x_test = x_test.reshape(-1, 28*28).astype("float32") / 255.0
```

```python
# Building the neural network
model = keras.Sequential([
    layers.Dense(128, activation="relu", input_shape=(784,)),  # hidden layer
    layers.Dense(64, activation="relu"),                       # hidden layer
    layers.Dense(10, activation="softmax")                     # output layer (10 classes)
])
```

```python
model.compile(optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])
```

```python
# Model training
history = model.fit(x_train, y_train, epochs=5, batch_size=32, validation_split=0.1, verbose=2)
```

```python
# training accuracy
train_acc = history.history['accuracy'][-1]
print(f"\nFinal Training Accuracy: {train_acc*100:.2f}%")
```

10. You are working on a project for a bank that wants to automatically detect
fraudulent transactions. The dataset is large, imbalanced, and contains structured features like transaction amount, merchant ID, and customer location. The goal is to
classify each transaction as fraudulent or legitimate.
Explain your real-time data science workflow:

‚óè How would you design a deep learning model (perceptron or multilayer NN)?

‚óè Which activation function and loss function would you use, and why?

‚óè How would you train and evaluate the model, considering class imbalance?

‚óè Which optimizer would be suitable, and how would you prevent overfitting?

1. Overall system & data pipeline (real-time constraints)

Feature store & streaming ingestion ‚Äî collect raw events (transaction amount, merchant_id, location, timestamp, device_id, etc.) into a streaming pipeline (Kafka/Kinesis) and materialize online features in a feature store for low-latency lookup (e.g., rolling counts, time-decayed averages, last transaction time).

Batch feature engineering for training ‚Äî compute historical windowed features (1h, 24h, 30d counts, averages, velocity features) and label with ground-truth fraud flags using a consistent time cutoff (no look-ahead). Use a time-based split for train/validation/test.

Real-time feature computation ‚Äî lightweight transforms (scaling, categorical encoding, feature lookups) performed at inference time with the same code used during training (feature parity).

Monitoring & feedback loop ‚Äî streaming metrics (prediction rates, latency), label ingestion for confirmed frauds, periodic model retraining, and drift detection.

2. Model design (architecture)

For structured tabular transaction data a multilayer perceptron (MLP) with careful handling of categorical features is a good default. If you have transactional sequences per user, consider a hybrid that adds a small temporal module (1D CNN / lightweight RNN / Transformer block) on top of per-transaction embeddings.

Recommended MLP architecture (low-latency, high-signal):

Inputs

Continuous features ‚Üí normalize (std or RobustScaler).

High-cardinality categorical features (merchant_id, device_id) ‚Üí learned embeddings (embedding dim ‚âà min(50, 1.5√ósqrt(cardinality))). Embeddings are compact and fast.

Low-cardinality categorical ‚Üí one-hot or small embeddings.

Binary flags / indicators as-is.

Network

Concatenate continuous inputs + flattened embeddings.

Dense(256) ‚Üí BatchNorm ‚Üí Activation ‚Üí Dropout(0.2)

Dense(128) ‚Üí BatchNorm ‚Üí Activation ‚Üí Dropout(0.2)

Dense(64) ‚Üí BatchNorm ‚Üí Activation

Output Dense(1) ‚Üí Sigmoid

Design notes

Keep model small (few layers, modest neuron counts) for real-time latency.

Use BatchNorm only if batch sizes during training are reasonable; LayerNorm is an alternative for small batches.

Prefer a single probabilistic output (sigmoid) for downstream thresholding and cost-sensitive decisions.

3. Activation & loss function (why)

Hidden activations: ReLU (or Leaky ReLU) ‚Äî fast, sparse gradients, stable training. If dying ReLU is a concern, use Leaky ReLU or ELU.

Output activation: Sigmoid, because we want a calibrated probability for thresholding and cost analysis.

Loss function choices (important for imbalance):

Binary Cross-Entropy (BCE) with class weights ‚Äî simple and effective: weight positive (fraud) examples higher to compensate imbalance.

Focal Loss ‚Äî if positives are extremely rare and you want to focus learning on hard positive examples, focal loss down-weights easy negatives and concentrates gradient on difficult/predictive frauds.

Reasoning: BCE + class weighting is standard and integrates well with calibration and thresholding. Focal loss can help in cases where negatives dominate gradient signal and many negatives are easy to classify.

4. Handling class imbalance (training & evaluation)

Training strategies

Class weighting in the loss (preferred for large datasets): set weight_pos = N_neg/N_pos or tune via validation to reflect business cost.

Focal loss (tunable gamma) if class weights alone are insufficient.

Careful resampling only when needed:

Undersampling negatives can speed training but risks losing rare patterns.

Oversampling positives (SMOTE etc.) may be useful for classical models; for deep nets prefer weighting or data augmentation (noise, small perturbations) over synthetic sampling that may distort features.

Batch composition: ensure each minibatch contains some positives (stratified or sampled) so gradients from fraud examples appear frequently and training is stable.

Evaluation metrics

Use metrics appropriate for imbalance ‚Äî Precision, Recall, F1, and especially Precision-Recall AUC (AUPRC) (more informative than ROC-AUC when positives are rare).

Business metrics: false positive rate (customer friction & operational cost) and false negative (undetected fraud monetary loss). Report cost-weighted metrics (expected monetary loss).

Threshold selection: pick threshold by maximizing business utility on validation set (not simply 0.5). Use expected cost function to choose operating point.

Temporal validation: use time-based splits (train on older data, validate on later period) to avoid leakage and mimic production distribution. Consider rolling origin cross-validation for robustness.

5. Optimizer & training recipe

Optimizer: Adam or AdamW (Adam with decoupled weight decay). AdamW provides better generalization when weight decay is used. Use default betas (0.9, 0.999) and tune learning rate.

Learning rate scheduling: cosine annealing or ReduceLROnPlateau; use warmup if training from scratch.

Batch size: medium batches (128‚Äì1024) depending on dataset and hardware ‚Äî ensure stable BatchNorm statistics or use LayerNorm.

Gradient clipping: clip by norm (e.g., 1.0‚Äì5.0) to prevent instability from rare large gradients.

Early stopping: monitor validation AUPRC or business metric with patience to stop training once no improvement.

Suggested hyperparameters (starting point):

AdamW, lr=1e-3, weight_decay=1e-5

Batch size 512 (tune), epochs 10‚Äì50 with early stopping

Class weight: pos_weight = N_neg / N_pos (then tune)

6. Preventing overfitting & improving robustness

Regularization

Weight decay (L2) via optimizer (AdamW).

Dropout (0.1‚Äì0.3) in dense layers.

Batch/Layer normalization to stabilize training.

Model capacity control: keep network small; prune useless neurons; prefer fewer parameters to limit memorization.

Data augmentation & noise: add small stochastic noise to numerical features to improve generalization; random masking for embeddings to reduce reliance on single high-cardinality IDs.

Feature selection & engineering: remove leaky features that reveal label or future info; cap outliers and winsorize skewed features.

Ensemble & calibration: use small ensembles (averaging 3 models) for stability but be mindful of inference cost. Calibrate predicted probabilities (Platt scaling or isotonic) using a holdout set to ensure probabilities reflect true risk.

Adversarial / robustness tests: test against simulated attack patterns (stolen cards, unusual merchant combos) and create training augmentation for those scenarios.

7. Deployment & operational considerations

Latency plan: keep model footprint small (< few MBs) and use inference-optimized runtimes (TF-Serving, ONNX, TorchScript). Aim for <50‚Äì100 ms tail latency.

Feature latency: precompute features that require heavy aggregation; use streaming aggregates with TTL and approximate counting (HyperLogLog/CountMin) if needed.

Fallback rules: always include deterministic rule-based overrides (e.g., block on confirmed blacklisted card) for safety.

Human-in-loop & feedback: route high-uncertainty predictions to manual review; feed confirmed labels back to training data for continuous learning.

Monitoring: model performance drift, feature distribution drift, prediction latency, false positive/negative rates, and business KPIs. Automate alerts and scheduled retraining.

8. Summary

Use a compact MLP with embeddings for categorical IDs; ReLU hidden units + Sigmoid output.

Use BCE with class weighting or Focal Loss to address severe imbalance.

Train with AdamW, LR scheduling, batch composition that ensures positives appear in minibatches, and early stopping.

Evaluate with Precision/Recall/F1 and AUPRC plus business cost metrics using time-based splits.

Prevent overfitting with weight decay, dropout, normalization, and keep the model small for low latency. Monitor production closely and iterate with a feedback loop.
