1. What is a Support Vector Machine (SVM)?

Support Vector Machine (SVM) is a supervised machine learning algorithm used for classification and regression tasks. It aims to find the optimal hyperplane that best separates data into different classes. The hyperplane is selected such that the margin‚Äîthe distance between the hyperplane and the nearest data points from each class (support vectors)‚Äîis maximized. SVMs are particularly effective in high-dimensional spaces and are capable of handling linearly and non-linearly separable data using the kernel trick. It focuses on the data points that are most difficult to classify, thereby achieving high accuracy and robustness against overfitting, especially in text classification, image recognition, and bioinformatics.

2. What is the difference between Hard Margin and Soft Margin SVM?

Hard Margin SVM assumes that data is perfectly linearly separable and constructs a hyperplane with no allowance for misclassification. It maximizes the margin while ensuring all training samples are correctly classified. However, it is highly sensitive to noise and outliers. Soft Margin SVM, in contrast, allows for some misclassification by introducing slack variables. This approach balances maximizing the margin and minimizing the classification error. A regularization parameter (C) controls the trade-off between the two. Soft Margin SVM is more practical for real-world data, which often contains noise and overlap between classes.

5. What are Support Vectors in SVM?

Support Vectors are the data points that lie closest to the decision boundary (hyperplane) in an SVM. They are critical because they directly influence the position and orientation of the hyperplane. Only these points are used to define the margin, while the rest of the dataset does not affect the boundary. Support vectors are the most informative data samples, particularly in sparse datasets. Their role also contributes to SVM's computational efficiency during prediction, as the model only needs to consider these critical points rather than the entire training set.

6. What is a Support Vector Classifier (SVC)?

Support Vector Classifier (SVC) is a specific implementation of SVM used for classification tasks. It attempts to find the best hyperplane that separates classes in the feature space, either linearly or using a kernel function for non-linear separation. In practice, sklearn.svm.SVC is widely used, offering parameters like C for regularization and kernel for specifying the transformation (linear, RBF, polynomial, etc.). SVC generalizes well in high-dimensional spaces and is particularly effective in tasks like text classification, image recognition, and bioinformatics where data may not be linearly separable.

7. What is a Support Vector Regressor (SVR)?

Support Vector Regressor (SVR) is an adaptation of SVM for regression problems. Instead of finding a hyperplane that separates data, SVR fits a function within an Œµ-tube (epsilon-insensitive zone) where deviations are not penalized. The objective is to find a function that has at most Œµ deviation from the actual outputs, and at the same time, is as flat as possible. Slack variables are introduced to handle points outside the Œµ-margin, with a regularization parameter controlling the trade-off between model complexity and the amount up to which deviations larger than Œµ are tolerated.

8. What is the Kernel Trick in SVM?

The Kernel Trick allows SVMs to perform non-linear classification by mapping input features into a higher-dimensional space where a linear separator might exist. Instead of computing the transformation explicitly, a kernel function computes the dot product of feature vectors in the transformed space directly. Common kernels include the polynomial kernel, radial basis function (RBF), and sigmoid. This technique enables SVMs to handle complex data distributions with computational efficiency and flexibility, making them powerful tools in pattern recognition and anomaly detection.

Linear Kernel: Best for linearly separable data; fast and interpretable.

Polynomial Kernel: Captures interactions between features by raising the dot product to a power

The choice of kernel depends on data complexity, dimensionality, and interpretability requirements.

10. What is the effect of the C parameter in SVM?

The C parameter in SVM controls the trade-off between maximizing the margin and minimizing the classification error. A high C value forces the algorithm to fit the training data more closely by penalizing misclassifications heavily, potentially leading to overfitting. A low C value allows a larger margin and more misclassifications, promoting generalization but possibly underfitting. Choosing the right C requires cross-validation and depends on the nature of the data‚Äînoisy datasets usually benefit from a lower C.

11. What is the role of the Gamma parameter in RBF Kernel SVM?

Gamma defines how far the influence of a single training example reaches. A low gamma implies a wider influence (i.e., smoother decision boundary), while a high gamma implies a narrow influence (complex, more localized boundary). In the RBF kernel, gamma controls the shape of the decision region. High gamma can cause overfitting by adapting too closely to the training data, whereas low gamma may lead to underfitting. Proper tuning using techniques like grid search and cross-validation is essential for optimal performance.

12. What is the Na√Øve Bayes classifier, and why is it called "Na√Øve"?

Na√Øve Bayes is a probabilistic classifier based on Bayes‚Äô Theorem, assuming conditional independence between features given the class label. It calculates the posterior probability of a class based on prior probabilities and likelihoods of features. It's called ‚Äúna√Øve‚Äù because it simplifies the modeling by assuming all features are independent, which rarely holds in practice. Despite this, it performs surprisingly well, especially in high-dimensional settings like text classification, spam detection, and sentiment analysis due to its efficiency and robustness.

14. Explain the differences between Gaussian, Multinomial, and Bernoulli Na√Øve Bayes:

Gaussian Na√Øve Bayes: Assumes features follow a normal distribution; suitable for continuous data.

Multinomial Na√Øve Bayes: Used for count-based features, such as word frequencies; ideal for text classification.

Bernoulli Na√Øes Bayes: Handles binary/boolean features; considers whether a feature is present or not.
Each variant makes different assumptions about feature distributions, guiding their applicability to different data types and domains.

15. When should you use Gaussian Na√Øve Bayes over other variants?

Gaussian Na√Øve Bayes is best suited for datasets with continuous numerical features that resemble a normal distribution, such as age, income, or temperature. It's commonly used in medical diagnosis, sensor data analysis, and other domains where features are continuous rather than categorical. If the assumption of normality holds reasonably well, Gaussian NB is efficient and interpretable. It's not ideal for sparse or count-based text data, where multinomial or Bernoulli variants are more appropriate.

16. What are the key assumptions made by Na√Øve Bayes?

Na√Øve Bayes assumes:

Conditional Independence: All features are independent given the class label.

Feature Distribution: Specific distributions (e.g., Gaussian, multinomial, or Bernoulli) model the likelihood of features.
These assumptions simplify the computation of class probabilities, making the model scalable and efficient. However, real-world data often violates the independence assumption. Still, Na√Øve Bayes often performs well due to its low variance and resilience to irrelevant features.

17. What are the advantages and disadvantages of Na√Øve Bayes?

Advantages:

Fast and computationally efficient

Works well with high-dimensional data

Performs well despite its strong assumptions
Disadvantages:

Assumes feature independence

Poor performance with correlated or complex features

Struggles with rare events in small datasets
Na√Øve Bayes is ideal for applications like text classification and spam filtering but may underperform in more intricate data relationships.

18. Why is Na√Øve Bayes a good choice for text classification?

Text data is typically high-dimensional and sparse, with features representing word counts or term presence. Na√Øve Bayes, especially the multinomial variant, is well-suited for this because:

It scales linearly with feature size.

Word occurrences in text are assumed conditionally independent, aligning with its core assumption.

It's robust to irrelevant features and works well even with small training data.
Its simplicity and speed make it a strong baseline for tasks like sentiment analysis, spam detection, and topic categorization.

19. Compare SVM and Na√Øve Bayes for classification tasks:

SVM excels in complex decision boundaries, is robust to overfitting (with proper regularization), and handles high-dimensional continuous data well.

Na√Øve Bayes is faster, requires less training data, and works well on discrete data like word counts.
SVM often yields higher accuracy, but Na√Øve Bayes is easier to interpret and deploy. Use SVM for complex, high-precision tasks and Na√Øve Bayes for fast, large-scale, or baseline models.

```python
# Write a Python program to implement a Na√Øve Bayes classifier for spam detection using email data
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

url = "https://raw.githubusercontent.com/justmarkham/pycon-2016-tutorial/master/data/sms.tsv"
df = pd.read_csv(url, sep='\t', header=None, names=['label', 'text'])

# Encode labels: 'ham' = 0, 'spam' = 1
df['label'] = df['label'].map({'ham': 0, 'spam': 1})

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['label'], test_size=0.3, random_state=42)

# Convert text to TF-IDF features
vectorizer = TfidfVectorizer(stop_words='english')
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

# Train Na√Øve Bayes model
model = MultinomialNB()
model.fit(X_train_tfidf, y_train)

# Predict
y_pred = model.predict(X_test_tfidf)

# Evaluate
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("Accuracy Score:", accuracy_score(y_test, y_pred))
```
